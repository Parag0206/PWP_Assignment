```python
import xarray as xr
import numpy as  np
import matplotlib.pyplot as plt
import PWP 
import PWP_helper as pwf
```

### Bay of Bengal

```python
# change the forcings 
normal = xr.open_dataset("input_data/forcings_85e15n.nc")
hf = xr.open_dataset("input_data/forcings_85e15n_new.nc")

hf["precip"] = normal["precip"]
hf["qlat"] = normal["qlat"]

hf.to_netcdf("input_data/final.nc")
```

```python
def rms(data):
    length = len(data)
    value = data**2
    value = value.sum()
    value = value/length
    value = value**0.5
    print(value)
    return value
```

```python
# Add to the mean 
total = normal

total["qlat"] = normal["qlat"]
total["precip"] = normal["precip"]
total["tx"] = rms(normal["tx"]) + hf["tx"]
total["ty"] = rms(normal["ty"]) + hf["ty"]
total["sw"] = normal["sw"].mean() + hf["sw"]
total["lw"] = normal["lw"].mean() + hf["lw"]
total["qsens"] = normal["qsens"].mean() + hf["qsens"]

# Make dataset
total.to_netcdf("input_data/rms_forcings.nc")
```

```python
def bob_run1():
    forcing_fname = 'forcings_85e15n.nc'
    prof_fname = 'initial_condition_85e15n.nc'
    print("Running Case with data from Bay of Bengal...")
    p={}
    p['rkz']=1e-6
    p['dz'] = 1.0
    p['max_depth'] = 400.0 
    p['rg'] = 0.25
    p['rb'] = 0.65
    p['lat'] = 15.
    p['winds_ON'] = True
    p['emp_ON'] = True
    p['heat_ON'] = True
    p['drag_ON'] = True 

    suffix = 'bob_run1'
    forcing, pwp_out = PWP.run(met_data=forcing_fname, prof_data=prof_fname, suffix=suffix, save_plots=True, param_kwds=p)
```

```python
bob_run1()
```

## EXP2 RUN1 : Free convection

```python
forcing_file_bob = xr.open_dataset("input_data/forcings_85e15n.nc")
hf_forcing_file = xr.open_dataset("input_data/rms_forcings.nc")

bob_1 = xr.open_dataset("output/pwp_output_exp2_run1.nc")
bob_2 = xr.open_dataset("output/pwp_output_hf_e2r1.nc")

bob_1 = bob_1.sel(time=slice('266.', '267.'))
bob_2 = bob_2.sel(time=slice('266.', '267.'))

plt.figure(figsize=(24,36))
f, (a0, a1, a2) = plt.subplots(3, 1, gridspec_kw={'height_ratios':[1.25,1.25,3]})
f.subplots_adjust(hspace=1)

forcing_file_bob["sw"].sel(time=slice('266.', '267.')).plot(ax=a0)
hf_forcing_file["sw"].sel(time=slice('266.', '267.')).plot(ax=a0)
a0.set_xticks(ticks=np.arange(266.,267.125,0.125), labels=[])
a0.set_yticks(ticks=[0.,500.])
a0.set_title("Shortwave")
a0.grid(axis='x')

tx = forcing_file_bob["tx"].sel(time=slice('266.', '267.'))
ty = forcing_file_bob["ty"].sel(time=slice('266.', '267.'))
tx_hf = hf_forcing_file["tx"].sel(time=slice('266.', '267.'))
ty_hf = hf_forcing_file["ty"].sel(time=slice('266.', '267.'))
a1.quiver(tx.time.values,0,tx,ty,color='b',
               units='y',
               scale_units='y',
               scale = 2,
               width=0.008,
               alpha=0.5)
a1.quiver(tx.time.values,0,tx_hf,ty_hf,color='y',
               units='y',
               scale_units='y',
               scale = 2,
               width=0.008,
               alpha=0.5)
a1.set_xticks(ticks=np.arange(266.,267.125,0.125), labels=[])
a1.set_yticks(ticks=[-0.05,0.,0.05], label="N/m2")
a1.set_title("Wind stress")
a1.grid(axis='x')

bob_1["mld"].plot(ax=a2, label="Actual", subplot_kws={"xlabel":"None"})
bob_2["mld"].plot(ax=a2, label="H/F", subplot_kws={"xlabel":"None"})

plt.title("Free convection")
plt.grid(axis='both', which="major")
plt.ylabel("MLD")
a2.legend(fontsize='x-small', bbox_to_anchor=(0.8,1.0))
plt.ylim(75,0)

f.suptitle("Bay of Bengal/JJAS", fontsize=12.)
plt.savefig("bob_e2r1_sep.pdf")
```
