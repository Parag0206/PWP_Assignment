# Mixed layer dynamics experiment 

## Region is Bay of Bengal: 85 $\degree$ E and 15 $\degree$ N

We carry out following experiments:
* Experiment 1:
    * Run 1 : All fluxes are ON
    * Run 2 : Freshwater fluxs are OFF
* Experiment 2(Mixing Experiment):
    * Run 1 : No wind 
    * Run 2 : Gradient mixing only
    * Run 3 : Bulk mixing only
    * Run 4 : Full model/all mixing schemes


```python
import numpy as np
import seawater as sw
import matplotlib.pyplot as plt
import PWP
from datetime import datetime
import PWP_helper as phf
import xarray as xr
```

# EXPERIMENT 1

## Experiment 1, Run 1 :  All Fluxes are ON

```python
def exp1run1():
    forcing_fname = 'forcings_85e15n.nc'
    prof_fname = 'initial_condition_85e15n.nc'
    print("Running Case with data from Bay of Bengal...")
    p={}
    p['rkz']=1e-6
    p['dz'] = 1.0
    p['max_depth'] = 400.0 
    p['rg'] = 0.25
    p['rb'] = 0.65
    p['lat'] = 15.
    p['winds_ON'] = True
    p['emp_ON'] = True
    p['heat_ON'] = True
    p['drag_ON'] = True 

    suffix = 'exp1_run1'
    forcing, pwp_out = PWP.run(met_data=forcing_fname, prof_data=prof_fname, suffix=suffix, save_plots=True, param_kwds=p)
```

```python
exp1run1()
```
