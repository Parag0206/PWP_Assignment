```python
import xarray as xr
import numpy as  np
import matplotlib.pyplot as plt
import PWP 
import PWP_helper as pwf
```

### Arabian Sea

```python
# change the forcings 
normal = xr.open_dataset("input_data/forcings_70e15n.nc")
hf = xr.open_dataset("input_data/forcings_70e15n_new.nc")

hf["precip"] = normal["precip"]
hf["qlat"] = normal["qlat"]

hf.to_netcdf("input_data/final.nc")
```

```python
def rms(data):
    length = len(data)
    value = data**2
    value = value.sum()
    value = value/length
    value = value**0.5
    print(value)
    return value
```

```python
# Add to the mean 
total = normal

total["qlat"] = normal["qlat"]
total["precip"] = normal["precip"]
total["tx"] = rms(normal["tx"]) + hf["tx"]
total["ty"] = rms(normal["ty"]) + hf["ty"]
total["sw"] = normal["sw"].mean() + hf["sw"]
total["lw"] = normal["lw"].mean() + hf["lw"]
total["qsens"] = normal["qsens"].mean() + hf["qsens"]

# Make dataset
total.to_netcdf("input_data/rms_forcings.nc")
```

```python
def as_run1():
    forcing_fname = 'forcings_70e15n.nc'
    prof_fname = 'initial_condition_70e15n.nc'
    print("Running Case with data from Arabian Sea...")
    p={}
    p['rkz']=1e-6
    p['dz'] = 1.0
    p['max_depth'] = 400.0 
    p['rg'] = 0.25
    p['rb'] = 0.65
    p['lat'] = 15.
    p['winds_ON'] = True
    p['emp_ON'] = True
    p['heat_ON'] = True
    p['drag_ON'] = True 

    suffix = 'as_run1'
    forcing, pwp_out = PWP.run(met_data=forcing_fname, prof_data=prof_fname, suffix=suffix, save_plots=True, param_kwds=p)
```

```python
as_run1()
```
